**1. Import Liblaries**

```python
import pandas as pd #Data Manipulation
import numpy as np  #Mathematical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.linear_model import LinearRegression #Train and Modeling
```

```python
dt = pd.read_csv('Lesson_4_student_grades_dataset.csv')
```

**Visualizing the Data**

```python
plt.scatter(dt['Previous_Exam_Scores'], dt ['Grades'])
plt.xlabel('Hours of Study')
plt.ylabel('Attendance Percentage')
plt.title('Relationship between Hours of Study and Attendance Percentage')
plt.show()
```

**Preparing Data for Modeling:**

```python
x = dt[['Previous_Exam_Scores']]
y = dt[['Grades']]
```

**Fitting the Linear Regression Model:**

```python
model = LinearRegression()
model.fit(x,y)
```

**Getting Model Coefficients:**

```python
slope = model.coef_[0]
intercept = model.intercept_
print('Slope',slope)
print('Intercept',intercept)
```

**Visualizing the Linear Regression Line:**

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='violet')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title('Scatter Plot of Data')
plt.show()
```

**Making Prediction**

```python
new_X = [[np.random.randint(100,200)]]
predicted_Y = model.predict(new_X)
print("Value of X:", new_X)
print("value of Y for new X:", predicted_Y)
```
