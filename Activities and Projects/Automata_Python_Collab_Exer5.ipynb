```python
#Import the DFA class from the automata.fa.dfa module
from automata.fa.nfa import NFA



# Create the NFA using the provided transitions
nfa_example = NFA(
    states={'q0', 'q1', 'q2'},
    input_symbols={'0', '1'},
    transitions={
    'q0': {'0': {'q0', 'q1'}, '1': {'q0'}},
    'q1': {'0': set(), '1': {'q2'}},
    'q2': {'0': set(), '1': set()},
    },
    initial_state='q0',
    final_states={'q2'}
)

# Check if a string is accepted
def is_accepted_example(input_str):
    return nfa_example.accepts_input(input_str)


test_strings = ['0101', '10010', '010101', '111', '0101010101']


for test_str in test_strings:
    if is_accepted_example(test_str):
        print(f'String "{test_str}" is accepted.')
    else:
        print(f'String "{test_str}" is not accepted.')
```

**B.NFA to DFA Conversion**

```python
# Imports the DFA class
from automata.fa.dfa import DFA

# Define DFA, states, transition, start, final, input symbols
dfa = DFA(
    states={'q0', 'q0q1', 'q0q2'},
    input_symbols={'0', '1'},
    transitions={
        'q0': {'0': 'q0q1', '1': 'q0'},
        'q0q1': {'0': 'q0q1', '1': 'q0q2'},
        'q0q2': {'0': 'q0q1', '1': 'q0'},
    },
    initial_state='q0',
    final_states={'q0q2'}
)

def is_accepted(input_str):
    return dfa.accepts_input(input_str)

test_strings = ["1101", "1000","11101","111", "1001"]

for test_str in test_strings:
    if is_accepted(test_str):
        print(f'String "{test_str}" is accepted.')
    else:
        print(f'String "{test_str}" is not accepted.')
```
