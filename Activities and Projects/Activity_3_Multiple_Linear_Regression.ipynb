# 1. Import Liblaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# 2. Importing Dataset

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
```

# Creating Dataframe

```python
x = pd.DataFrame(data, columns = ["Hours_of_Study","Attendance_Percentage","Previous_Exam_Scores","Extracurricular_Activities","Travel_Time_to_School",])

y = data["Grades"]

data.head()
```

```python
# @title Attendance_Percentage

from matplotlib import pyplot as plt
data['Attendance_Percentage'].plot(kind='line', figsize=(8, 4), title='Attendance_Percentage')
plt.gca().spines[['top', 'right']].set_visible(False)
```

# 4. Splitting Dataset to Training and Testing

```python
x_train, x_test, y_train, y_test = train_test_split(data,y, test_size=0.2, random_state=42)
```

# 5. Initializing the Model

```python
model = LinearRegression()
```

# 6. Fitting the Model

```python
model.fit(x_train, y_train)
```

# 7. Making Predictions

```python
y_predict = model.predict(x_test)
print(x_test)
print (y_predict)
```

# 8. Calculation Mean Squared Error

```python
mse = mean_squared_error(y_test,y_predict)
print(f"Mean Squared Error", mse)
```

# 9. New Grade Prediction

```python
new_grade = [[8.79, 86, 63, 1, 50, 86]]
predicted_grades = model.predict(new_grade)
print(f"Predicted Grade : {predicted_grades[0]}")
```
