# 1. Import Liblaries

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

# 2. Load Datasets

```python
data = pd.read_csv('monthly-beer-production-in-austr.csv')
```

# 3. Convert Month to datetime

```python
data['Month'] = pd.date_range(start='1956-01',periods=len(data),freq='M')
data.set_index('Month',inplace=True)
```

# 4. Plot the Time Series

```python
plt.figure(figsize=(10,6))
plt.plot(data.index,data['Monthly beer production'], marker='o', linestyle= 'solid')
plt.title('Productivity of Beer')
plt.xlabel('Month')
plt.ylabel('Production')
plt.xticks(rotation=20)
plt.grid(True)
plt.show()
```

# 5. Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(data['Monthly beer production'], model = '-')
fig=decomposition.plot()
plt.show()
```

# 6. Forecasting with Arma

```python
from statsmodels.tsa.stattools import adfuller

adf_test = adfuller (data['Monthly beer production'])

print ('ADF Statistics:', adf_test[0])
print ('p-value: ', adf_test[1])
```

# 7. Fit an Arima Model

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(data['Monthly beer production'], order=(1, 1, 1))
```
